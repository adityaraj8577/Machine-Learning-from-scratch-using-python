## KDE(kernel density estimation)

```python
#### allows us to estimate pdf from our finite datasets (non_parametric way)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

### Univariate(1D)

```python
sns.kdeplot(cars.horsepower);
```

```python
sns.kdeplot(cars.horsepower,fill =True)
```

```python
sns.kdeplot(cars.horsepower,bw=10);
```

```python
sns.kdeplot(cars.horsepower , cumulative =True)
```

### Bivariate plot

```python
sns.kdeplot(x= cars.horsepower,y=cars.mpg)
```

```python
sns.kdeplot(x = cars.horsepower, y= cars.mpg, n_levels = 20)
```

```python
sns.kdeplot(x =cars.horsepower, y= cars.mpg,shade = True)
```

```python
sns.set_style('whitegrid')
sns.kdeplot(x =cars.horsepower,y=cars.mpg , shade=True, shade_lowest = False)
```

```python
sns.kdeplot(x =cars.horsepower,y = cars.mpg, shade =True, shade_lowest =False,cbar= True)
```

### Seaborn Distplot(Distribution plot)

```python
# Distplot = Histogram + kdeplot+ rugplot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars  =sns.load_dataset('mpg').dropna()
```

```python
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('white')
```

```python
sns.distplot(cars.weight)  # default distplot = hist+kdeplot
```

```python
sns.distplot(cars.weight, kde= False)
```

```python
sns.distplot(cars.weight,rug = True,hist = False)
```

```python
sns.distplot(cars.weight, bins=6)
```

```python
sns.distplot(cars.weight,bins=[1500,2000,3500,5200])
```

### Fit

```python
from scipy.stats import norm
```

```python
sns.distplot(cars.weight,fit = norm,kde =False)
```

```python
from scipy.stats import skewnorm
sns.distplot(cars.weight,fit = skewnorm, kde =False)
```

### kwargs hist_kws

```python
# alpha is used for the brightness by default = 0.5
```

```python
sns.distplot(cars.weight, kde = False,hist_kws={'alpha':1})
```

```python
sns.distplot(cars.weight,kde=False, hist_kws={'alpha':1 , 'histtype': 'stepfilled'})
```

## kde_kws

```python
sns.distplot(cars.weight, kde_kws={'lw':5}, hist=False)
```

```python
sns.kdeplot(cars.weight,lw=5)
```

```python
sns.distplot(cars.weight,kde_kws={'bw':6})
```

## rug_kws

```python
sns.distplot(cars.weight, rug = True)
```

```python
sns.distplot(cars.weight, rug= True, color='r')
```

```python
sns.distplot(cars.weight, rug =True, rug_kws={'height':0.1,'color':'r'})
```

# BOXPLOT

```python
# Box having 25th percentile, median,75th percentile
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars  =sns.load_dataset('mpg').dropna()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
# Filter down to only cars with even numbers of cylinders
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.boxplot(cars.mpg)
```

```python
cars.mpg.describe()
```

```python
sns.boxplot(x=cars.origin, y =cars.mpg)
```

```python
# Another way of writing the same
```

```python
sns.boxplot(x='origin',y='mpg',data=cars)
```

# Hue

```python
sns.boxplot(x='origin',y='mpg',hue='cylinders',data=cars)
```

```python
cars.model_year.describe()
```

```python
cars['newer_model']= cars.model_year>76
```

```python
sns.boxplot(x='origin',y='mpg',hue='newer_model',data=cars)
```

```python
sns.boxplot(x='mpg',y='origin',data=cars)
```

```python
# if want to change the ordering
sns.boxplot(x='mpg',y='origin',data=cars,order=['japan','europe','usa'])
```

```python
sns.boxplot(x='mpg',y='origin',hue='newer_model',data=cars, order =['japan','europe','usa'],hue_order=[True,False])
```

```python
sns.boxplot(x='mpg',y='origin',data=cars,color='g')
```

```python
sns.boxplot(x='mpg',y='origin',hue='newer_model',data=cars,color='g')
```

```python
sns.boxplot(x='mpg',y='origin',data=cars,width=0.1)
```

```python
sns.boxplot(x='mpg',y='origin',data= cars,linewidth=2.5)
```

```python
sns.boxplot(x='mpg',y='origin',data=cars,whis=1) # whis the length of horizontal line
```

```python
sns.boxplot(x='mpg',y='origin',data=cars,fliersize=20)
```

```python
sns.boxplot(x='mpg',y='origin',data=cars, showcaps=False)
```

# VIOLIN PLOT

```python
# categorical distribution plot
# Violin  = kde two faces  + boxplot
# violin gives pdf i.e. kde plot


```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
# filtering down to only cars with even number of cylinders
```

```python
cars =  cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(cars.displacement)
```

```python
sns.violinplot(x='cylinders',y='displacement',data=cars)
```

```python
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars)
```
