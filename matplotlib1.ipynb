### Random dots in a scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sizes = (30*np.random.rand(N))*2 # 0 to 15 point radii
plt.scatter(x,y,s=sizes,c=colors,alpha=0.5)
plt.show()
```

### Subplots,part1

#### 221 = top left
#### 222 = top right
#### 223 = bottom left
#### 224 = bottom right

```python
fig = plt.figure()
fig.suptitle('My Subplots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222, facecolor = 'y')
fig.add_subplot(223)
fig.add_subplot(224)
plt.show()
```

### subplots part2

```python
fig,plots = plt.subplots(2,sharex =True)
fig.suptitle('Sharing X axis')
x = range(0,200,5)
y = [n**0.8 for n in x]
plots[0].plot(x,y,color='r')
plots[1].scatter(x,y)
plt.show()
```

### Save figure to image file

```python
plt.figure(figsize =(4,3),dpi = 100)
plt.plot([245,150,220,300,20])
plt.axis([0,7,0,300])
plt.title('Flight Data')
plt.xlabel('Speed')
plt.savefig('Flights.png')
plt.show()
```

### Creating stem graph

```python
x = np.linspace(0.1,2*np.pi,41)
y = np.exp(np.sin(x)) 
plt.stem(x, y, use_line_collection = True) 
plt.show() 
```

```python
# data to display on plots 
x = [1, 2, 3, 4, 5, 6, 7, 4] 
# This will plot a simple histogram
plt.hist(x, bins = [1, 2, 3, 4, 5, 6, 7])
# Title to the plot
plt.title("Histogram")
# Adding the legends
plt.legend(["bar"])
plt.show()
```

### Stack plot

```python
days = [1,2,3,4,5]
studying =[7,8,9,6,11]
playing = [8,5,7,8,13]
plt.stackplot(days,studying,playing,colors =['r','c'])
plt.xlabel('days')
plt.ylabel('No. of hours')
```
