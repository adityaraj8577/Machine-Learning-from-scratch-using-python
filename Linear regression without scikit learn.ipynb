```python
import pandas as pd

penguins = pd.read_csv("penguins.csv")
penguins
```

```python
a=penguins.loc[:,['flipper_length_mm','body_mass_g']]
a.head()
```

```python
a.head()
```

```python
import seaborn as sns
feature_name = 'flipper_length_mm'
target_name = 'body_mass_g'
data = a[[feature_name]]  ## Double bracket is used because we want it to be a dataframe not a series
target = a[target_name]  ## single bracket is used because we want it to be series
plot=sns.scatterplot(x='flipper_length_mm',y='body_mass_g',data=a,alpha=0.5)
plot.set_title('Body mass as a function of flipper length')
```

```python
# linearly varies hence we can say that y= a*x+b , a and b are parameters of our model
```

```python
def linear_model_flipper_mass(flipper_length,parameter_w,parameter_b):
    
    body_mass = parameter_w*flipper_length + parameter_b
    return body_mass
```

```python
# weight_flipper_length = 45
# intercept_body_mass = -5000
```

```python
import numpy as np
w = 45
b = -5000
flipper_length_range = np.linspace(data.min(),data.max(),num=300)
predicted_body_mass = linear_model_flipper_mass(flipper_length_range,w,b)
```

```python
label = '{0:.2f} (g/mm) * flipper length + {1:.2f} (g)'
```

```python
plot_s= sns.scatterplot(x='flipper_length_mm',y='body_mass_g',data=a,alpha=0.5,color='black')
plot_s.plot(flipper_length_range,predicted_body_mass)
plot_s.set_title(label.format(w , b))
label.format(w , b)
```

```python
#If the coefficient is negative, it means that penguins with shorter flipper lengths have larger body masses.
```
