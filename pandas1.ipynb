## How do I sort a pandas DataFrame or Series??

```python
movies1 = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies1.head()
```

```python
movies1.title.sort_values() #ONLY WE WILL GET IS THAT COLUMN
# also we can use movies['title'].sort_values()
# this way we get the data in  ascending order
# if want in descending then .sort_values(ascending = False)
```

```python
movies1['title'].sort_values(ascending = False) # we will get only this column
```

```python
movies1.sort_values('title') # WE WILL GET ALL THE COLUMNS 

```

```python
movies1.sort_values(['content_rating','duration']) # SORTING TWO VALUES AT A TIME
```

## how to filter row by column values

```python
import pandas as pd
```

```python
movies2= pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies2.head()
```

```python
movies2.shape
```

```python
type(True)
type(False)
```

```python
# APPLYING SOME CONDITION
booleans = []
for length in movies2.duration:
    if length>=200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
len(booleans)
```

```python
is_long = pd.Series(booleans) # method of converting list into pandas series  by using pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies2[is_long] # Note here movies('title') it will give title series but without inverted comma it will give the applied condition
# it will only show the data having True value
```

```python
#2ND METHOD
is_long = movies2.duration>=200
is_long.head()
```

```python
movies2[is_long]
```

```python
# Above step can also be written as
movies2[movies2.duration>=200]
```

```python
# What if we want to see only the genre section of the movies having >=200
movies2[movies2.duration>=200]['genre']
```

```python
# for BETTER practice we use .loc  means we want to sort on the basis of row and and also want to show a particular column
movies2.loc[movies.duration>=200,'genre']
```
