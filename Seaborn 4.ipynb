```python
import seaborn as sns

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.barplot(x='species',y='body_mass_g',data=penguins)

# Generally in barplot ine axis is categorical whilw other is numerical
```

```python
penguins.groupby('species').body_mass_g.mean()
```

```python
sns.barplot(x= 'body_mass_g',y ='species',data=penguins)
```

## Ci

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,ci=None)
```

## Estimator

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,estimator=np.std)
# Also wecan use different function like min.max.mean etc
```

## CAtegorical variable

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,hue='sex')
```

```python
sns.barplot(y='species',x='body_mass_g',data=penguins,hue='sex')
```

### order, Hue order

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,order=['Gentoo','Chinstrap','Adelie'])
```

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,order=['Gentoo','Chinstrap','Adelie'],
               hue_order =['Female','Male'],hue='sex')
```

```python
sns.set_style('white')
```

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,color='b')
```

### errcolor and errwidth

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,color='c',errcolor='yellow',errwidth=8)
```

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,palette='hot')
```

## Edge color and linewidth

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,ci=None,color='aliceblue', edgecolor='green',lw=3)
```

# COUNTPLOT

```python
# Barplot it gives the mean of the weight of the given variable and also ci
```

```python
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')

diamonds.head()
```

```python
diamonds = diamonds[diamonds.clarity.isin(['SI1', 'VS2'])]
```

```python
diamonds.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.countplot(x='color',data=diamonds)
```

```python
diamonds.color.value_counts(sort=False)
```

```python
sns.countplot(y='color',data=diamonds)
```

```python
sns.countplot(x='color', data=diamonds);
```

```python
sns.countplot(x='cut', data=diamonds);
```

```python
diamonds.dtypes
```

```python
diamonds.color.cat.categories
```

```python
diamonds.cut.cat.categories
```

```python
color_order=['J', 'I', 'H', 'G', 'F', 'E', 'D']

sns.countplot(x='color', data=diamonds, order=color_order);
```

```python
diamonds.color.value_counts().index
```

```python
sns.countplot(x='color',data=diamonds,order =diamonds.color.value_counts().index )
```

```python
sns.countplot(x='color',data=diamonds,order =diamonds.color.value_counts().index[::-1] )
```

```python
sns.countplot(x='color', data=diamonds, hue='clarity')

# we have filtertd but still it is showing all 
```

```python
sns.countplot(x='color', data=diamonds, hue='clarity', hue_order=['SI1', 'VS2'])
```

```python
sns.countplot(x='color', data=diamonds, color='skyblue');
```
