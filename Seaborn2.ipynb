# LINEPLOT

```python
import seaborn as sns
from matplotlib import pyplot as plt

import pandas as pd
from datetime import datetime
```

```python
df= pd.read_csv('dataset.csv', parse_dates=[3])
```

```python
df.head(10)
```

```python
# Do a bit of data cleaning
# Rename some columns
# Create three new date/time columns
# filter down to only two garage locations
```

```python
df.rename(columns={'SystemCodeNumber':'Location' , 'LastUpdated':'Timestamp'},inplace=True)
df['Day']= df.Timestamp.dt.date
df['Month']= df.Timestamp.dt.month
df['Hour'] = df.Timestamp.dt.hour
```

```python
park = df[df.Location.isin(['Broad Street','NIA South'])]
```

```python
park.head()
```

```python
# BAsics
```

```python
sns.set_style('dark')
```

```python
months = [datetime(2016,10,1),datetime(2016,11,1),datetime(2016,12,1)]
plt.rc('date.autoformatter', day='%b %Y')
```

```python
sns.lineplot(x='Day',y='Occupancy',data=park)
plt.xticks(months)
```

```python
sns.lineplot(x='Hour',y='Occupancy',data=park)
```

## Bootstrapping

```python
 sns.lineplot(x='Hour',y='Occupancy',data=park,n_boot = 1000)
```

```python
# n_boot basically gives the number of bootstrap samples
```

```python
sns.lineplot(x='Hour',y='Occupancy',data=park,n_boot=10)
```

```python
sns.lineplot(x='Hour',y='Occupancy',data=park,ci=95)

# ci is basically the confidence interval 
```

```python
# To turn off bootstrap ci set ci=None

sns.lineplot(x='Hour',y='Occupancy',data=park,ci=None)
```

```python
# bootstrap we get is on the basis of mean , if we want to change then
sns.lineplot(x='Hour',y='Occupancy',data=park,estimator ='sum')
```

```python
sns.lineplot(x='Hour',y='Occupancy',data=park,estimator ='std')
```

# Visual Semantics

```python
sns.lineplot(x='Day',y='Occupancy',data=park,hue='Location')
plt.xticks(months)
```

```python
sns.lineplot(x='Day',y='Occupancy',data=park,hue='Location',palette = ['gray','xkcd:brick red'])
```

```python
# style is used just like the hue is used but if gives different representation ie in dotted line
```

```python
sns.lineplot(x='Day',y='Occupancy',data=park,style='Location')
```

```python
sns.lineplot(x='Day',y='Occupancy',data=park,hue='Location',style='Location')
```

```python
sns.lineplot(x='Hour',y='Occupancy',data=park,size= 'Location')

# Difference betwwen size and style 
# size = it give the another plot with the same linetype
# style  = it gives the another plot with the different linetype
```

```python
sns.lineplot(x='Hour',y='Occupancy',data=park,ci=None,size='Month')
```

```python
# Most of the matplotlib line styling works within Seaborn lineplot
```

```python
sns.lineplot(x='Day',y='Occupancy',data=park,ci=None, lw=4, color = '#aa00aa',alpha=0.5)
```
